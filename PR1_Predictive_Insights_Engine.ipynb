# PART A: CONCEPTUAL UNDERSTANDING

## 1. What are Supervised Learning Algorithms?

Supervised Learning Algorithms are machine learning algorithms that learn from labeled data.  
The model is trained using input features and known output values so it can predict results for new unseen data.

Examples:
- Linear Regression
- Logistic Regression
- Decision Tree
- Random Forest

---

## 2. Difference Between Regression and Classification Algorithms

| Regression Algorithms | Classification Algorithms |
|----------------------|---------------------------|
| Predict continuous numerical values | Predict categorical labels |
| Output is numeric | Output is class/category |
| Example: House Price Prediction | Example: Spam Email Detection |
| Algorithms: Linear Regression | Algorithms: Logistic Regression |

Example:
- Regression → Predict house price = ₹45,00,000
- Classification → Predict email = Spam or Not Spam

---

## 3. Explain Simple Linear Regression

Simple Linear Regression is a supervised learning algorithm used to predict a continuous value using one independent variable and one dependent variable.

It tries to find the best-fit straight line between input and output.

Equation:

Y = mX + c

Where:
- Y = Dependent Variable
- X = Independent Variable
- m = Slope
- c = Intercept

Example:
- Predicting house price using house area.

---
## Assumptions of Linear Regression

### 1. Linear Relationship
The relationship between input (X) and output (Y) should be linear.

#### Example:
If study time increases, marks should increase in a straight-line pattern.

---

### 2. Independence
The data points should be independent of each other.

### Example:
One student's marks should not affect another student's marks.

---

### 3. No Multicollinearity
Input variables should not be highly related to each other.

#### Example:
"Age in years" and "Age in months" together can create multicollinearity.

---

### 4. Homoscedasticity
The error should have constant variance for all data points.

#### Simple Meaning:
Errors should be spread equally.

---

### 5. Normal Distribution of Errors
The errors (residuals) should follow a normal distribution.

#### Simple Meaning:
Most errors should be small and balanced around zero.

---

## Simple Table

| Assumption | Simple Meaning |
|---|---|
| Linear Relationship | Data follows straight-line relation |
| Independence | Data points are unrelated |
| No Multicollinearity | Inputs should not be highly connected |
| Homoscedasticity | Equal error spread |
| Normal Distribution | Errors should be normally distributed |

---
## 5. What is Bias-Variance Trade-Off?

Bias-Variance Trade-Off is the balance between underfitting and overfitting.

- High Bias → Model is too simple → Underfitting
- High Variance → Model is too complex → Overfitting

A good model maintains a balance between bias and variance for better prediction accuracy.

---

## 6. Explain Overfitting and Underfitting

## 1. Underfitting
Underfitting happens when the model is too simple and cannot learn the data properly.

### Characteristics:
- High Bias
- Low Variance

### Meaning:
- **High Bias:** Model makes many mistakes because it learns very little.
- **Low Variance:** Model gives almost similar predictions for all data.

### Result:
- Poor training accuracy
- Poor testing accuracy

### Example:
A student studies very little and fails in both practice and final exams.

---

## 2. Overfitting
Overfitting happens when the model learns training data too much, including noise and small details.

### Characteristics:
- Low Bias
- High Variance

### Meaning:
- **Low Bias:** Model learns training data very well.
- **High Variance:** Model changes a lot with new data and gives unstable predictions.

### Result:
- High training accuracy
- Low testing accuracy

### Example:
A student memorizes answers exactly but cannot answer new questions.

---

# 3. Best Fit (Balanced Model)

### Characteristics:
- Low Bias
- Low Variance

### Result:
- Good training accuracy
- Good testing accuracy
- Works well on new data

---

# Simple Table

| Concept | Bias | Variance | Performance |
|---|---|---|---|
| Underfitting | High | Low | Poor |
| Overfitting | Low | High | Poor |
| Best Fit Model | Low | Low | Good |

---

# Part G: GRADIENT DESCENT OPTIMIZATION

### 21. Explain Gradient Descent Conceptually

- Gradient Descent is an optimization algorithm used to minimize the cost function and improve model accuracy.

- The algorithm repeatedly updates model parameters to reduce prediction error.

##### Working:
- 1. Start with random parameter values.
- 2. Calculate prediction error.
- 3. Compute gradients.
- 4. Update parameters.
- 5. Repeat until minimum error is reached.

#### Formula:
θ = θ − α ∇J(θ)

Where:
- θ = Parameters
- α = Learning Rate
- ∇J(θ) = Gradient of Cost Function

---

### 22. Implement Batch Gradient Descent from scratch.

````python
import time
X = df['area_sqft'].values
y = df['house_price_inr'].values
````

````python
X = (X - X.mean()) / X.std()
````

````python
def batch_gradient_descent(X, y, learning_rate, epochs):
    w = 0
    b = 0

    m = len(X)

    cost_history = []
    start = time.time()

    for i in range(epochs):

        y_pred = w * X + b
        error = y_pred - y
        dw = (1/m) * np.sum(error * X)
        db = (1/m) * np.sum(error)

        w = w - learning_rate * dw
        b = b - learning_rate * db

        cost = (1/(2*m)) * np.sum(error**2)
        cost_history.append(cost)

    end = time.time()
    training_time = end - start
    return w, b, cost_history, training_time
````

````python
w_batch, b_batch, batch_cost_history, batch_time = batch_gradient_descent(
    X,
    y,
    learning_rate=0.01,
    epochs=100
)
````

````python
print("Weight :", w_batch)
print("Bias :", b_batch)
print("Final Cost :", batch_cost_history[-1])
print("Training Time :", batch_time)
````

### 23. Stochastic Gradient Descent (SGD).

````python
def stochastic_gradient_descent(X, y, learning_rate, epochs):

    w = 0
    b = 0
    m = len(X)
    cost_history = []
    start = time.time()
    for epoch in range(epochs):
        for i in range(m):

            index = np.random.randint(m)
            x_sample = X[index]
            y_sample = y[index]

            y_pred = w * x_sample + b
            error = y_pred - y_sample
            dw = error * x_sample
            db = error

            w = w - learning_rate * dw
            b = b - learning_rate * db

        total_pred = w * X + b
        cost = (1/(2*m)) * np.sum((total_pred - y)**2)
        cost_history.append(cost)

    end = time.time()
    training_time = end - start
    return w, b, cost_history, training_time
````

````python
w_sgd, b_sgd, sgd_cost_history, sgd_time = stochastic_gradient_descent(
    X,
    y,
    learning_rate=0.01,
    epochs=100
)
````

````python
print("Weight :", w_sgd)
print("Bias :", b_sgd)
print("Final Cost :", sgd_cost_history[-1])
print("Training Time :", sgd_time)
````

### 24.Mini-Batch Gradient Descent

````python
def mini_batch_gradient_descent(X,
                                y,
                                learning_rate,
                                epochs,
                                batch_size):
    w = 0
    b = 0
    m = len(X)
    cost_history = []
    
    start = time.time()
    for epoch in range(epochs):
        indices = np.random.permutation(m)
        X_shuffle = X[indices]
        y_shuffle = y[indices]

        for i in range(0, m, batch_size):
            X_batch = X_shuffle[i:i+batch_size]
            y_batch = y_shuffle[i:i+batch_size]

            y_pred = w * X_batch + b
            error = y_pred - y_batch
            
            dw = (1/len(X_batch)) * np.sum(error * X_batch)
            db = (1/len(X_batch)) * np.sum(error)

            w = w - learning_rate * dw
            b = b - learning_rate * db

        total_pred = w * X + b
        cost = (1/(2*m)) * np.sum((total_pred - y)**2)
        cost_history.append(cost)

    end = time.time()
    training_time = end - start
    return w, b, cost_history, training_time
````

````python
w_mini, b_mini, mini_batch_cost_history, mini_batch_time = mini_batch_gradient_descent(
    X,
    y,
    learning_rate=0.01,
    epochs=100,
    batch_size=32
)
````

````python
print("Weight :", w_mini)
print("Bias :", b_mini)
print("Final Cost :", mini_batch_cost_history[-1])
print("Training Time :", mini_batch_time)
````

````python
print("Batch GD Time :", batch_time)

print("SGD Time :", sgd_time)

print("Mini-Batch GD Time :", mini_batch_time)
````
